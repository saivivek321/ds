```python
#1- Python program to print all prime numbers between 1 to 1000
import math
def isprime(N):   #function to find whether a given number is prime or not
  if(N==1):
    return False
  for i in range(2,(int)(math.sqrt(N)+1)):  
    if(N%i==0):                   #if we find any divisor of N except 1 and N then returh false
      return False
  return True
print('Prime Numbers between 1 and 1000 are :- ')
for i in range(1,1001):   #checking every number in range between 1 and 1000
  if(isprime(i)):
    print(i,end=" ")
```

```python
#2 - Bubble sort
def Bubble(A):
    n = len(A)
    for i in range(n-1):     # Traverse the array elements
      for j in range(0, n-i-1):   # traverse the array from 0 to n-i-1 , according to bubble sort algo Last i elements are already at the right place
        if(A[j] > A[j + 1]):     # Swap if the element found is greater
          temp = A[j]
          A[j] = A[j+1]
          A[j+1] = temp
      print("Pass # " , i+1 ," : ", end = " ")
      print(A)

N = (int)(input("Enter the Number of elements to sort:- "))  
print("Enter the array elements:- ")
A = []
for i in range(N):
  el = input()
  A.append(el)
print("Array Before Sorting : ", A)
Bubble(A)
print("Sorted elements of given array are:- ",A)

           
                
```

```python
#3 - Python program to compute the sum of two matrices.
def SUM(A,B):
  M = len(A)
  N = len(A[0])
  S = []

  for i in range(len(A)):    #Interate through the rows 
    row = []
    for j in range(len(A[0])):  #Interate through the columns
      row.append(A[i][j] + B[i][j])     #find the sum of A[i][j] and B[i][j]
    S.append(row)
  return S

r = int(input("Enter the number of rows:"))
c = int(input("Enter the number of columns:"))

A = []
print("Enter the matrix A values in rowwise order:")
for i in range(r):          # A for loop for row entries
  row = []
  for j in range(c):      # A for loop for column entries
    row.append((int(input())))
  A.append(row)

B = [] 
print("Enter the matrix B values in rowwise order:")
for i in range(r):          # A for loop for row entries
  row = []
  for j in range(c):      # A for loop for column entries
    row.append((int(input())))
  B.append(row)
C = SUM(A,B)
# For printing the sum of both matrices
print("Sum of given two matrix is:- ")
for i in range(r):
    for j in range(c):
        print(C[i][j], end = " ")
    print()
```

```python
#Question - 4
def recursive_binary_search(A,low,high,E):
  if(low>high):
    return -1
  mid = (int)((low+high)/2)
  if(A[mid]==E):
    return mid
  elif A[mid] < E:
    return recursive_binary_search(A,mid+1,high,E)
  else:
    return recursive_binary_search(A,low,mid-1,E)
  
def Iterative_binary_search(A,E):
  n = len(A)
  low = 0
  high = n-1
  while(low<high):
    mid = (int)((low+high)/2)
    if(A[mid]==E):
      return mid
    elif A[mid] < E:
      low = mid+1
    else:
      high = mid-1
  return -1

A = []
n = int(input("Enter the number of elements in list:- "))
print('Enter the elements in the sorted order:- ')
for i in range(n):
  el = int(input())
  A.append(el)
F = int(input("Enter the element to search in the given array:- "))
a = recursive_binary_search(A,0,n-1,F)  # calling the recusive binary search function
if(a!=-1):
  print("Element ",F," is present at the index :- " ,a);
else:
  print("Element ",F," is not present in the given array");

b = Iterative_binary_search(A,F)    # calling the Iterative binary search function
if(b!=-1):
  print("Element ",F," is present at the index :- " ,b);
else:
  print("Element ",F," is not present in the given array");
```
