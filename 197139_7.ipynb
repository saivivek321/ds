```python
import pandas as pd                
import numpy as np
from sklearn.preprocessing import StandardScaler  # For Standardizing the Data

import keras
from keras.models import Sequential   #For building the Neural Network Model layer by layer
from keras.layers import Dense        # Groups linear stack of layers and used to add fully connected layer in ANN

from sklearn.metrics import accuracy_score            # to determine the accuracy score
```

```python
dataset= pd.read_csv('insurance_data.csv')           
dataset
```

```python
# Number of Columns
n=dataset.shape[1]    
# Number of Rows           
m=dataset.shape[0]  

# Dividing Dataset into Training and Testing (80% -20%)
x_train=np.array(dataset.iloc[:int(0.8*m),:n-1])      
y_train=np.array(dataset.iloc[:int(0.8*m),-1])         
x_test=np.array(dataset.iloc[int(0.8*m):,:n-1])
y_test=np.array(dataset.iloc[int(0.8*m):,-1])        


# Standardizing the training and testing data using standardscaler()
SC = StandardScaler()

x_train_scaled= SC.fit_transform(x_train)   
x_test_scaled=SC.fit_transform(x_test)   

model=Sequential()                                      # Create Neural Network Model 
model.add(Dense(2, input_dim=2, activation='relu'))     # Input Dimension in this layer is 2 and we use reLU Activation Function
model.add(Dense(2, input_dim=2, activation='relu'))     # Input Dimension in this layer is 2 and we use reLU Activation Function
model.add(Dense(1, activation='softmax'))               # Output dimension has only 1 parameter(0 or 1) and we use softmax activation function  
                                                     

# Adam is the optimizer that can perform the gradient descent and accelerate the gradient descent algorithm 
# loss function as binary_crossentropy as we have 2 cases only 0 or 1                                                        
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])


model.fit(x_train_scaled, y_train, epochs=100,batch_size=25) # Fit the model 
model.evaluate(x_test_scaled,y_test)                        # Evalvuate using scaled testing data
y_pred=model.predict(x_test_scaled)                         # Predict based on scaled testing data


#Calculating Accuracy of the Model
actual=y_test.tolist()                                 
prediction=y_pred.tolist()                                   
a = accuracy_score(prediction,actual)                     # Used accuracy_score function from sklearn.metrics
print('Accuracy of the Model is :', a*100)
```

```python
import pandas as pd                
import numpy as np
from sklearn.preprocessing import StandardScaler  # For Standardizing the Data
```

```python
# importing the dataset
dataset= pd.read_csv('Churn_Modelling.csv')           
dataset
```

```python
# Drop 1st 3 Columns as we don't need them
print("Drop RowNumber ,CustomerID and Surname from Data")
dataset.drop(dataset.columns[[0,1,2]],axis=1,inplace=True)   
dataset
```

```python
# Number of Columns
n=dataset.shape[1]    
# Number of Rows           
m=dataset.shape[0]  


def Convert_text_to_int(Array):
  visited=[]
  res=[]
  index=0
  for i in Array:
    if(i in visited):
      res.append(visited.index(i))
    else:
      visited.append(i)
      res.append(index)
      index=index+1
  return res

# converting the text data in geography and gender column into Integer Data
dataset.iloc[:,1]=np.array(Convert_text_to_int(dataset.iloc[:,1].tolist()))     
dataset.iloc[:,2]=np.array(Convert_text_to_int(dataset.iloc[:,2].tolist()))        
dataset

```

```python
# Dividing Dataset into Training and Testing (80% -20%)
x_train=np.array(dataset.iloc[:int(0.8*m),:n-1])      
y_train=np.array(dataset.iloc[:int(0.8*m),-1])         
x_test=np.array(dataset.iloc[int(0.8*m):,:n-1])
y_test=np.array(dataset.iloc[int(0.8*m):,-1])        


# Standardizing the training and testing data using standardscaler()
SC = StandardScaler()

x_train_scaled= SC.fit_transform(x_train)   
x_test_scaled=SC.fit_transform(x_test)   



model=Sequential()                                      # Create Neural Network Model 
model.add(Dense(10, input_dim=10, activation='relu'))   # Input dimensions to this layer is 10 and we use reLU activation function
model.add(Dense(6, input_dim=10, activation='relu'))    # input dimensions to this layer is 8 and we use reLU activation function 
model.add(Dense(4, input_dim=6, activation='relu'))     # input dimensions to this layer is 6 and we use reLU activation function 
model.add(Dense(1, activation='sigmoid'))               # Output dimension has only 1 parameter so we use sigmoid activation function  
                                                       

# Adam is the optimizer that can perform the gradient descent and accelerate the gradient descent algorithm 
# loss function as binary_crossentropy as we have 2 cases only 0 or 1                                                        
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])


# Fit the Model
model.fit(x_train_scaled, y_train, epochs=100,batch_size=250) 
model.evaluate(x_test_scaled,y_test)                         # Evaluate using our scaled testing data
y_pred=model.predict(x_test_scaled)                          # Predict based on scaled testing data


#Calculating Accuracy of the Model
actual=y_test.tolist()                                     
prediction=y_pred.tolist()                                     
for i in range(len(prediction)):                    # since we used sigmoid function 
  if(prediction[i][0]>=0.5):                        # we convert those values into either 0 or 1 
    prediction[i]=1                                 # if greater or equal to 0.5 we make it 1 
  else:                                            
    prediction[i]=0                                 # if less than 0.5 we make it equal to 0 

a = accuracy_score(prediction,actual)                     
print('Accuracy of Model is :', a*100)            
```
